# Collocation recognition in academic Russian

```
import pandas as pd
import os
```

## Legal domain

```
Law_4gr = pd.read_csv('d3_4grams_cvalued.csv', delimiter='\t', quotechar='`')


Law_4gr = Law_4gr.sort_values('d3_pmi', ascending=False).dropna()
Law_4gr.to_csv(os.path.join('/content/drive/My Drive/Диплом про коллокации', 'law_4gr_pmi.csv'))
```

### 4grams sorted by PMI descending

```
Law_4gr.sort_values('d2_pmi', ascending=False).head(10)
```

4grams sorted by PMI with c-value > 1.2

```
Law_4gr['d2_c-value'] = pd.to_numeric(Law_4gr['d2_c-value']) / Law_4gr['d2_freq']
```

```
Law_4gr_c_val_1_2 = Law_4gr[Law_4gr['d2_c-value'] > 1.2].copy()
```

```
Law_4gr_c_val_1_2.to_csv(os.path.join('/content/drive/My Drive/Диплом про коллокации', 'Law_4gr_c_val_1_2.csv'))
Law_4gr_c_val_1_2.head()
```

### Filtered by c-value and sorted by logDice

```
Law_4gr_c_val_logDice = Law_4gr_c_val_1_2.sort_values('d2_logdice', ascending=False)
Law_4gr_c_val_logDice.to_csv(os.path.join('/content/drive/My Drive/Диплом про коллокации', 'Law_4gr_c_val_logDice.csv'))
Law_4gr_c_val_logDice.head(10)
```

Filtered by c-val and sorted by t-score

```
Law_4gr_c_val_tsc = Law_4gr_c_val_1_2.sort_values('d2_t-score', ascending=False)
Law_4gr_c_val_tsc.to_csv(os.path.join('/content/drive/My Drive/Диплом про коллокации', 'Law_4gr_c_val_tsc.csv'))
Law_4gr_c_val_tsc.head(10)
```

### Combined method

```
Law_4gr_c_val_1_2['PMI_rank'] = range(Law_4gr_c_val_1_2.shape[0])
```

```
Law_4gr_c_val_1_2['d2_logdice'].sort_values(ascending=False).index
rank = pd.Series(range(Law_4gr_c_val_1_2.shape[0]))
rank.index = Law_4gr_c_val_1_2['d2_logdice'].sort_values(ascending=False).index
```

```
Law_4gr_c_val_1_2['logDice_rank'] = rank
```

```
Law_4gr_c_val_1_2['d2_t-score'].sort_values(ascending=False).index
rank = pd.Series(range(Law_4gr_c_val_1_2.shape[0]))
rank.index = Law_4gr_c_val_1_2['d2_t-score'].sort_values(ascending=False).index

Law_4gr_c_val_1_2['t-score_rank'] = rank
```

```
Law_4gr_c_val_1_2.head()
```

```
Law_4gr_c_val_1_2['combined_rank'] = Law_4gr_c_val_1_2['t-score_rank'] + Law_4gr_c_val_1_2['logDice_rank']*0.5 + Law_4gr_c_val_1_2['PMI_rank']*0.7
```

```
Law_4gr_combined = Law_4gr_c_val_1_2.sort_values('combined_rank')
Law_4gr_combined.to_csv(os.path.join('/content/drive/My Drive/Диплом про коллокации', 'Law_4gr_combined.csv'))
Law_4gr_combined.head(10)
```

### POS-filter

```
def pos_filter(df, columns):
    filter = ['PUNCT', 'SYM', 'X']
    filtered = df[~df[columns].isin(filter).any(axis=1)]
    return filtered
```

```
Law_4gr_comb_pos = pos_filter(Law_4gr_combined, ['pos1', 'pos2', 'pos3', 'pos4'])
Law_4gr_comb_pos.head(10)
```

```
Law_4gr_comb_pos.to_csv(os.path.join('/content/drive/My Drive/Диплом про коллокации', 'Law_4gr_comb_pos.csv'))
```
